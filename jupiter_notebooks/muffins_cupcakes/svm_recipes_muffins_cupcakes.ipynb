# Support Vector Machine (Muffins vs. Cupcakes)

```python
# Data analysis
import numpy as np
import pandas as pd
from sklearn import svm

# Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2)

%matplotlib inline
```

```python
recipes = pd.read_csv('svm_recipes_muffins_cupcakes.csv')
recipes
```

```python
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={'s': 70})
```

We need to format or pre-process our data

```python
type_label = np.where(recipes['Type'] == 'Muffin', 0, 1)
type_label
```

```python
recipe_features = recipes.columns.values[1:].tolist()
recipe_features
```

```python
ingredients = recipes[['Flour', 'Sugar']].values
ingredients
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(ingredients, type_label)
```
