# Activation Functions

Also known as **Transfer Functions**, activation functions can be linear (a.k.a. Identity) and non-linear. The non-linear type is the most used because it helps with generalisation.

Sigmoid, tanh, Softmax, ReLU, Leaky ReLU

```python
import seaborn as sns
import numpy as np
```

## Binary Step Function

It depends on a threshold that decides whether the neuron should be activated or not.

$$
f(x) =
\begin{cases}
    0 & x \lt 0 \\
    1 & x \geq 0 \\
\end{cases}
$$

### Disadvantages
1. The derivative (gradient) is zero everywhere, except at $x=0$ where it's non-differentiable.
1. It can only be used with binary classification

The first disadvantage is a dealbreaker. Backpropagation is all about differentiability. Even if it were differentiable for every $x \in R$, having a gradient always zero would cause the gradient descent mechanism to stop immediately.

```python
def binary_step_function(x):
    return np.where(x < 0, 0, 1)
    
x = np.linspace(-10, 10, 100)
y = binary_step_function(x)

sns.lineplot(x=x, y=y);
```

## Linear Activation Function

Also known as **Identity Function**, the output of this function is equal to the input.

$$
f(x) = x
$$

```python
def linear_activation_function(x):
    return x
    
x = np.linspace(-10, 10, 100)
y = linear_activation_function(x)

sns.lineplot(x=x, y=y);
```
