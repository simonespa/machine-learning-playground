To solve a system of equations in multiple variable, we can use the addition/subtraction method to generate a new equation in one variable, then solve for that variable and then plug the value of that solved variable to solve in the other variable.

For example, given a system of equations like the following

- $2y - 3x = 7$
- $5x = 4y - 12$

Rearrange the equations

- $-3x + 2y = 7$
- $5x - 4y = - 12$

Multiply the first equation by $2$ to make the coefficient of $y$ opposite to the coefficient of $y$ of the second equation

- $-6x + 4y = 14$
- $5x - 4y = - 12$

We can now subtract the two equations to cancel $y$

- $-x = 2 \implies x = -2$

Now that we have found the value for $x$, we can plug it into one of the two original equations of the system to find the value of $y$

$X = -2$:
- $12 + 4y = 14 \implies 4y = 2 \implies y = \frac{1}{2}$
- $-10 - 4y = -12 \implies 4y = 2 \implies y = \frac{1}{2}$

## Local/Global Maxima/Minima

```python
from sympy import *
from sympy.plotting import plot
```

```python
x, y = symbols('x, y')
```

```python
f = x**3 -3 * x**2 - 9 * x - 1
f
```

```python
diff_f = diff(f)
diff_f
```

```python
solve(diff_f)
```

```python
f.subs({x: -1})
```

```python
f.subs({x: 3})
```

```python
plot(f, diff_f);
```
