```python
module_id = 'savasy/bert-base-turkish-ner-cased'
ner = pipeline('ner', model=module_id, tokenizer=module_id)
```

```python
ner("Benim adım Simone ve Londra'da yaşıyorum.") # My name is Simone and I live in London.
```

 # Questions and Answers

```python
bert_qa = BertForQuestionAnswering.from_pretrained('bert-base-uncased')
bert_qa
```

```python
module_id = 'deepset/roberta-base-squad2'
qa = pipeline('question-answering', model=module_id, tokenizer=module_id, revision='v1.0')
```

```python
qa(question="Where do you live now?", context="I live in London.")
```

```python
qa(question="Where do you live now?", context="I like beer.")
```
