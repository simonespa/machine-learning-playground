# BERT flavours

```python
from transformers import pipeline
```

```python
nlp = pipeline('fill-mask', model='bert-base-cased')
```

```python
type(nlp.model)
```

```python
MASK = nlp.tokenizer.mask_token
print(MASK)
preds = nlp(f"If you don't stop stealing, you will be in {MASK}.")
```

```python
for p in preds:
    print(f"Token: {p['token_str']} - Score: {p['score'] * 100:.2f}%")
```

```python
roberta_nlp = pipeline('fill-mask', model='roberta-base')
```

```python
MASK = roberta_nlp.tokenizer.mask_token
print(MASK)
roberta_preds = roberta_nlp(f"If you don't stop stealing, you will be in {MASK}.")
```

```python
for p in roberta_preds:
    print(f"Token: {p['token_str']} - Score: {p['score'] * 100:.2f}%")
```
