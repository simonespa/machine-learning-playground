```python
import torch
from transformers import BertForMaskedLM, pipeline
```

```python
bert_lm = BertForMaskedLM.from_pretrained('bert-base-cased')
```

```python
bert_lm
```

```python
nlp = pipeline(task='fill-mask', model='bert-base-cased') # or model=bert_lm
```

```python
type(nlp.model)
```

```python
nlp.tokenizer
```

```python
preds = nlp(f'If you do not {nlp.tokenizer.mask_token} at the sign, you will get a fine.')
```

```python
len(preds)
```

```python
for p in preds:
    print(f"Token: '{p['token_str']}' Score: {p['score'] * 100:.2f}%")
```
