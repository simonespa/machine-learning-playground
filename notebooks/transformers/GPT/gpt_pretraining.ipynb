# Pre-training GPT

```python
from transformers import pipeline, set_seed, GPT2Tokenizer
from torch import tensor
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
generator = pipeline('text-generation', model='gpt2', tokenizer=tokenizer)
```

```python
set_seed(31)
```

```python
generator('The holocaust was', max_length=10, num_return_sequences=10, temperature=0.8)
```

```python
generator(
    'The following text gives an example of sentiment analysis. If the text is "I hate losing at a game.", the sentiment is "Negative". If the text is "My life is great.", then the sentiment is "Positive". Now, if I give you the text "I love this new song", the sentiment is',
    top_k=2, temperature=0.1, max_length=55)[0]['generated_text']
```
