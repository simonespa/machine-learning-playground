```python
import statsmodels.api as sm
import pandas
from patsy import dmatrices
```

```python
df = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['Department', 'Lottery', 'Literacy', 'Wealth', 'Region']].dropna()
```

```python
df.head()
```

```python
df.info()
```

```python
y, X = dmatrices('Lottery ~ Literacy + Wealth + Region', data=df, return_type='dataframe')
```

```python
mod = sm.OLS(y, X)
```

```python
res = mod.fit()
```

```python
res.summary()
```
